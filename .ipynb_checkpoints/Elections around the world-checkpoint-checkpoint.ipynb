```python
import pandas as pd #pandas is a package designed for data analysis
import numpy as np #numpy is a great package that has a tonne of in-built statistical formulae
import matplotlib.pyplot as plt #matplotlib is a great plotting library
plt.style.use('fivethirtyeight') #this adds a customised style to my plots
% matplotlib inline 
```

```python
#all data is from here:
#http://www.parlgov.org/data/table/view_election/
#reading in my dataset
elex = pd.read_csv('http://www.parlgov.org/static/data/development-cp1252/view_election.csv', encoding='latin1')

#turning the election date information to a datetime format so we can do timeseries analysis
elex['election_date'] = pd.to_datetime(elex['election_date'
])
```

```python
#take a quick look at the data we have
elex.head()
```

```python
#creating a 'year' column using pandas datetime function
elex['year'] = elex['election_date'].dt.year

#plotting number of parties involved each year
elex['year'].value_counts().sort_index().plot(kind='line')
#that's a pretty useless graph
```

```python
#plotting the change in parliamentary vs. pres elections per year
elex.groupby('year').election_type.value_counts().unstack().plot(kind='line')
#what's ep??? that's not presidential. i should probably look into that. 
```

```python
#parties that have received the most seats in the UK 
elex[elex.country_name=='United Kingdom'].groupby('party_name').seats_total.sum().sort_values().plot(kind='bar', figsize=(15,15), fontsize=15)
```

```python
#what's the spectrum of left to right in the UK
elex[elex.country_name=='United Kingdom'].groupby('party_name').left_right.mean().sort_values().plot(kind='bar', figsize=(15,15),  fontsize=15)
```
